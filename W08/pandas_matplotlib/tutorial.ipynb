# Input

```python
import pandas as pd


# Load the DataFrame
df_1544 = pd.read_table('c:/Users/merho/Desktop/tutorial/1544.csv', delimiter=';', encoding='latin-1', na_values="****")
```

```python
df_1544.head()
```

# Convert the ,

```python
# Define all potential columns that could need conversion
columns_to_convert = [
    'Zeit  1 - Standardmessrate', 'WI 1',
    'WI 2', 'WI 3',
    'WI 4', 'WI 5', 'WI 6',
    'WI 7', 'Temperatur', 'WI 8',
    'DMS - innen (200 mm)', 'DMS - außen (350 mm)', 'PMX_1 CH 3-4', 'PMX_1 CH 4-1', 'PMX_1 CH 4-2',
    'PMX_1 CH 4-3', 'PMX_1 CH 4-4', 'PMX_1 dig in', 'PMX_1 dig out'
]

# Convert columns
for column in columns_to_convert:
    df_1544[column] = df_1544[column].astype(str).str.replace(',', '.').astype(float)

df_1544.head()
```

# Plot the figure

```python
import matplotlib.pyplot as plt

# Ensure your DataFrame df_1544 is properly defined and the necessary columns have been converted to floats.

# Plotting the data from df_1544
plt.figure(figsize=(13, 8))

# Plot various 'WI' columns against the 'Zeit  1 - Standardmessrate' column.
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 5'], label='V_innen_r', color='blue')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 8'], label='V_außen_r', color='green')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 7'], label='H_Kopf_r', color='red')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 6'], label='H_Fuß_r', color='cyan')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 4'], label='V_außen_l', color='magenta')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 1'], label='V_innen_l', color='yellow')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 2'], label='H_Kopf_l', color='black')
plt.plot(df_1544['Zeit  1 - Standardmessrate'], df_1544['WI 3'], label='H_Fuß_l', color='orange')

plt.title('Verschiebung über Zeit')
plt.xlabel("Zeit (s)")
plt.ylabel("Länge (mm)")
plt.legend()
plt.grid(True)

# Save the figure
plt.savefig('figure.png', dpi=300)

# Show it in Jupyter notebook
plt.show()
```

# Correct the offset errors

```python
# Filter the DataFrame for the first 10 seconds
filtered_df = df_1544[df_1544['Zeit  1 - Standardmessrate'] <= 10]

# Calculate the mean of the 'WI 3' column within the first 10 seconds and adjust the column
mean_wi3 = filtered_df['WI 3'].mean()
df_1544['WI 3'] = df_1544['WI 3'] - mean_wi3
print("Mean of 'WI 3' for the first 10 seconds in df_1544:", mean_wi3)

# Calculate the mean of the 'WI 7' column within the first 10 seconds and adjust the column
mean_wi7 = filtered_df['WI 7'].mean()
df_1544['WI 7'] = df_1544['WI 7'] - mean_wi7
print("Mean of 'WI 7' for the first 10 seconds in df_1544:", mean_wi7)
```
